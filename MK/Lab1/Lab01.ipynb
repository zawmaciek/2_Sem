```python
import numpy as np
import matplotlib.pyplot as plt
def sigmoid(x):
    return 1/ (1+np.exp(-x))
def swish(x,beta):
    return 2*x*sigmoid(beta*x)

x = np.linspace(-5, 5, 1000)

swish_gradient = np.diff(swish(x, 1)) / np.diff(x)

fig, ax = plt.subplots()
ax.plot(x[:-1], swish(x, 1)[:-1], label='Swish Function')
ax.plot(x[:-1], swish_gradient, label='Gradient of Swish Function')
ax.legend()
plt.show()
```

Funkcja aktywacji Swish zyskała popularność w nauczaniu maszynowym, ponieważ wykazano, że przewyższa inne popularne funkcje aktywacji, takie jak ReLU i sigmoid w niektórych scenariuszach. Oto kilka potencjalnych przypadków użycia funkcji aktywacji Swish:

Klasyfikacja obrazów: Swish może być używany jako funkcja aktywacji w konwencjonalnych sieciach neuronowych (CNN) do zadań klasyfikacji obrazów. Wykazano, że Swish osiąga lepsze wyniki niż ReLU w zadaniach klasyfikacji obrazów na dużą skalę.

Przetwarzanie języka naturalnego: Wykazano, że Swish poprawia wydajność w niektórych zadaniach przetwarzania języka naturalnego, takich jak modelowanie języka i analiza sentymentu.

Recurrent Neural Networks: Wykazano, że Swish osiąga dobre wyniki w rekurencyjnych sieciach neuronowych (RNN) w zadaniach modelowania sekwencji, takich jak rozpoznawanie mowy i tłumaczenie maszynowe.

Zadania regresji: Wykazano również, że Swish dobrze radzi sobie w zadaniach regresji, gdzie celem jest przewidywanie wartości ciągłej, a nie kategorycznej.

Ogólnie rzecz biorąc, Swish okazał się być solidną funkcją aktywacji, która może być używana w różnych architekturach sieci neuronowych i aplikacjach. Należy jednak pamiętać, że nie ma jednej uniwersalnej funkcji aktywacji, a wybór funkcji aktywacji powinien być oparty na konkretnym zadaniu i używanej architekturze.
