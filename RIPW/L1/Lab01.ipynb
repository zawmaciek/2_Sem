przedstawiÂ´c sposoby laËczenia ramek danych za pomocaË metod merge i
concat

```python
df1 = pd.DataFrame({'key': ['A', 'B', 'C', 'D'], 'value1': [1, 2, 3, 4]})
df2 = pd.DataFrame({'key': ['B', 'D', 'E', 'F'], 'value2': [5, 6, 7, 8]})
merged_df = pd.merge(df1, df2, on='key')
merged_df
```

```python
df1 = pd.DataFrame({'key': ['A', 'B', 'C', 'D'], 'value1': [1, 2, 3, 4]})
df2 = pd.DataFrame({'key': ['B', 'D', 'E', 'F'], 'value2': [5, 6, 7, 8]})
concatenated_df = pd.concat([df1, df2], axis=0)
concatenated_df
```

pokazaÂ´c dodawanie nowych kolumn za pomocaË operacji matematycznych

```python
merged_df['sum']=merged_df['value1']+merged_df['value2']
merged_df
```

 przedstawiÂ´c na przyk ladzie dodawanie nowych kolumn z pomocaË funkcji

```python
df = pd.DataFrame({'A': [1, 2, 3, 4], 'B': [5, 6, 7, 8]})
df['C'] = df.apply(lambda row: row['A'] + row['B'], axis=1)
df
```
